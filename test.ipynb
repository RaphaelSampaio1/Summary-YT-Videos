```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_video_id(url_yt):
    return url_yt.split("watch?v=")[-1]
```

```python
url = "https://www.youtube.com/watch?v=yKAuUAPREMw"
video_id = get_video_id(url)
```

```python
transcipt = YouTubeTranscriptApi.get_transcript(video_id, languages=['en','pt','es'])
```

```python
transcipt
```

## Merging the whole text

```python
transcipt_joined = " ".join([line['text'] for line in transcipt])
display(transcipt_joined)
```

## Using A.I to summary this context

```python
import g4f

def ask_gpt(messages: list) -> str:
    response = g4f.ChatCompletion.create(
        model=g4f.models.gpt_4o,
        messages=messages
    )
    print(response)
    return response

messages = [
    {"role": "user", "content": f"Resuma em tpicos essa transcrio do vdeo do YouTube na linguagem que estiver escrita:\n\n{transcipt_joined}"}
]

# === 4. Enviar ao modelo e mostrar o resultado ===
resposta = ask_gpt(messages)
print("\n Resumo em tpicos:\n")
print(resposta)
```
